```python
#!/usr/bin/python
from urllib.request import urlopen
from urllib import parse
from bs4 import BeautifulSoup

urlList = []



def getAllUrl(url):
    print(url)
    try:
        page = urlopen( url ).read()
    except:
        return []
    try:
        soup = BeautifulSoup(page)
        soup.prettify()
        for anchor in soup.findAll('a', href=True):
            if not 'http://' in anchor['href']:
                if parse.urljoin(url, anchor['href']) not in urlList:
                    urlList.append(parse.urljoin(url, anchor['href']))
            else:
                if anchor['href'] not in urlList:
                    urlList.append(anchor['href'])

        length = len(urlList)

        return urlList
    except (urllib2.HTTPError, e):
        print (e)

def hierurl(url):
    urls=getAllUrl(url)
    print("Erste Hierarchieebene geschafft")
    for url in urls:
        getAllUrl(url)
        
def listAllUrl(urls):
    for x in urls:
        print (x)
        urls.remove(x)
        urls_tmp = getAllUrl(x)
        for y in urls_tmp:
            urls.append(y)
            
url="http://www.sncf.com"
urls=hierurl(url)
```

```python
import urllib
from  nltk import RegexpTokenizer
from bs4 import BeautifulSoup

import nltk
from nltk.corpus import stopwords
stop = set(stopwords.words('french'))

import codecs
# ...

tokenizer = RegexpTokenizer(r'\w+')

def get_text(url):
    text=None
    if "#" not in url:
        
        html = urlopen(url).read()
        soup = BeautifulSoup(html.decode('utf-8'))

        # kill all script and style elements
        for script in soup(["script", "style"]):
            script.extract()    # rip it out

        # get text
        text = soup.get_text()

        # break into lines and remove leading and trailing space on each
        lines = (line.strip() for line in text.splitlines())
        # break multi-headlines into a line each
        chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
        # drop blank lines
        text = '\n'.join(chunk for chunk in chunks if chunk)
        tokens = tokenizer.tokenize(text)
        tokens=[i.lower() for i in tokens if i not in stop ]
        #tokens=[i for i in tokens if i not in stop]

    return(tokens)
```
